```python
import time
def countdown(t):
    while t>=0:
        m,s=divmod(t,60)
        t-=1
        print("Minutes: ",m,"| Seconds: ",s,end='\r')
        time.sleep(1)
    return "Fire in the hole."
t=int(input("Timer (IN SECONDS): "))
countdown(t)
```

```python
import random
import math

def begin():
    while True:
        x=input("Enter a command (Type 'HELP' to see commands): ")
        if x.lower()=='help':
            print("1. Type PLAY to play the roulette.")
            print("2. Type STATS to see your stats.")
            print("3. Type RULES to see the rules of the game.")
        
        elif x.lower()=='play':
            while True:
                money=float(input("How much money do you have: "))
                if money<100:
                    print("ERROR: MUST BE LESS THAN 100")
                else: test(money)
                
        elif x.lower()=='rules':
            print("\n1. In this game you will be betting a certain amount of money")
            print("2. The computer will choose a random number that you have to guess between 1 and 49")
            print("3. EVEN numbers have the color: BLACK")
            print("4. ODD numbers have the color: RED")
            print("5. If you guess the right number you win 3 times the amount you bet.")
            print("6. If not, we're going to check if your number's color matches the roll's color.")
            print("7. If the colors are matching, you will only loose half the amount you bet.")
            print("8. If they don't, you will loose all of your bet.")
        
        elif x.lower()=='stats':
            try:
                print("You currently have:",money)
            except:
                print("You have no money! Start the game to have stats.")
        else: print("Unknown command. Type 'HELP' to see commands")

def test(money):
    bet=float('inf')
    while True:
        bet=float(input("How much are you betting: "))
        if bet>money:
            print("MUST BET LESS THAN",money)
            play(money,bet)
        elif 1>=bet:
            print("MUST BET AT LEAST 1$")
        play(money,bet)

def gameEnd(money):
    print("WARNING: Type 'Continue' if you want to continue without resetting your money")
    while True:
        x=input("Give a command: ")
        if x.lower()=="play":
            begin()
        elif x.lower=="help":
            print("1. Type PLAY to play the roulette.")
            print("2. Type STATS to see your stats.")
            print("3. Type RULES to see the rules of the game.")
    
        elif x.lower()=="stats":
            print("You currently have:",money)
        elif x.lower()=="rules":
            print("\n1. In this game you will be betting a certain amount of money")
            print("2. The computer will choose a random number that you have to guess between 1 and 49")
            print("3. EVEN numbers have the color: BLACK")
            print("4. ODD numbers have the color: RED")
            print("5. If you guess the right number you win 3 times the amount you bet.")
            print("6. If not, we're going to check if your number's color matches the roll's color.")
            print("7. If the colors are matching, you will only loose half the amount you bet.")
            print("8. If they don't, you will loose all of your bet.")
        else: print("Unkown command. Type 'HELP' to see commands.")
def play(money,bet):
    myGuess=0
    while 1>myGuess or myGuess>49:
        myGuess=int(input("Guess a number between 1 and 49: "))
        if 1>myGuess or myGuess>49:
            print("ERROR: Must be between 1 and 49")
    myGuessDict={'value':'','color':''}
    myGuessDict['value']=myGuess
    if myGuess%2==0:
        myGuessDict['color']='BLACK'
    else: myGuessDict['color']='BLACK'

    roll=random.randint(1,50)
    rollDict={'value':'','color':''}
    rollDict['value']=roll
    print("\nThe ball landed on: ",rollDict['value'])
    if rollDict['value']%2==0:
        rollDict['color']='BLACK'
    else: rollDict['color']='RED'

            
    if myGuessDict['value']==rollDict['value']: #VALUE IS VALUE
        money+=bet*3
        print("Congratulations! You won: ",bet*3)

        #AFTER GAME
        gameEnd(money)
    
    elif myGuessDict['value']!=rollDict['value']: #VALUE NOT VALUE
        if myGuessDict['color']==rollDict['color']: #COLOR IS COLOR
            money-=bet/2
            print("Your guess' color: ",myGuessDict['color'])
            print("The rolled number's color: ",rollDict['color'])
            print("Bad luck! You lost: ",bet/2)
                
            #AFTER GAME
            gameEnd(money)

        elif myGuessDict['color']!=rollDict['color']: #COLOR NOT COLOR
            money-=bet
            print("Your guess' color: ",myGuessDict['color'])
            print("The rolled number's color: ",rollDict['color'])
            print("Bad luck! You lost: ",bet)
                
            #AFTER GAME
            gameEnd(money)
            
begin()
```
